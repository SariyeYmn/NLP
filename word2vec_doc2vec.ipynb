```python
from gensim.models import Word2Vec
import openpyxl
import re
from collections import Counter
```

```python
def cumleleri_kelimelere_ayir(cumle):
    # Cümlenin içindeki kelimeleri ayırmak için düzenli ifade kullanabilirsiniz
    kelimeler = re.findall(r'\b\w+\b', cumle)
    return kelimeler
```

```python


# Excel dosyasını açma
dosya_yolu = 'C:\\Users\\Sariye\\Desktop\\YZM-3\\NLP\\IMDBtop25_yorumlari(temizlenmis_hali).xlsx'
excel_wb = openpyxl.load_workbook(dosya_yolu)

# İlk sayfayı seçme
sayfa = excel_wb.active

# Kelimeleri saklamak için bir liste oluştur
tum_kelimeler = []

# Her satırdaki cümleleri kelimelere ayır ve listeye ekle
for satir in sayfa.iter_rows(values_only=True):
    kelime_listesi=[]
    for cumle in satir:
        kelimeler = cumleleri_kelimelere_ayir(cumle)
        kelime_listesi.extend(kelimeler)
    tum_kelimeler.append(kelime_listesi)
    
# Tüm kelimeleri düz bir liste haline getirme
tum_kelimeler_flat = [kelime for kelime_listesi in tum_kelimeler for kelime in kelime_listesi]

kelime_sayilari = Counter(tum_kelimeler_flat)
en_cok_tekrar_eden_kelimeler = kelime_sayilari.most_common(20)

print("En çok tekrar eden 20 kelime:")
for kelime, sayi in en_cok_tekrar_eden_kelimeler:
    print(f"{kelime}: {sayi}")

# Kelimeleri ekrana yazdır
print(tum_kelimeler)

# Dosyayı kapatma
excel_wb.close()
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Öncelikle, excel dosyanızı okuyun
df = pd.read_excel('C:\\Users\\Sariye\\Desktop\\YZM-3\\NLP\\IMDBtop25_yorumlari(temizlenmis_hali).xlsx')

# Veri çerçevesindeki metin sütununu alın
texts = df['body'].tolist()

# Her bir metni TaggedDocument formatına dönüştürün
tagged_data = [TaggedDocument(words=text.split(), tags=[str(i)]) for i, text in enumerate(texts)]
```

```python


# Doc2Vec modelini eğitin
model = Doc2Vec(vector_size=100, window=5, min_count=1, workers=4, epochs=20)
model.build_vocab(tagged_data)
model.train(tagged_data, total_examples=model.corpus_count, epochs=model.epochs)

# Örnek cümle listesi
sample_sentences = ["performance made smile",# 75.cümle
                    "relationship strong feel first moment make contact one another",#154.cümle
                    "hugely familiar tim robbins great job demanding role",#170 cümle
                    "never gives quest become free man",#199 cümle
                    "one best dramas time"#233 cümle
                    ]


# Her bir örnek cümle için benzer cümleleri bulun
for sample_sentence in sample_sentences:
    # Seçilen cümleyi vektöre dönüştürün
    sample_vector = model.infer_vector(sample_sentence.split())

    # Tüm cümle vektörlerini alın
    all_vectors = [model.dv[str(i)] for i in range(len(texts))]

    # Cosine similarity kullanarak benzer cümleleri bulun
    similarities = cosine_similarity([sample_vector], all_vectors)

    # En benzer olan 3 cümleyi bulun
    top_indices = similarities.argsort()[0][-4:-1]

    # Seçilen cümle ve en benzer cümlelerin indekslerini yazdırın
    print(f"\nSeçilen cümle: {sample_sentence}")
    print("En benzer cümleler:")
    for idx in top_indices:
        print(f"{idx +2}. cümle: {texts[idx]}")
```
