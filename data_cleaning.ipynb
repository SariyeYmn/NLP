```python
#kontrol etmek istediğim sütun adı
target_column = 'body'
```

```python
def numerical_value(target_column):
    # Nümerik değer içeren satırları bulma
    rows_with_numeric_values = data[data[target_column].str.contains('\d', na=False)]
    # Eğer rows_with_numeric_values boş değilse, o sütunda nümerik değerler bulunuyor demektir.
    if  not rows_with_numeric_values.empty:
       print(f"{target_column} sütununda nümerik değerler bulunuyor.")
       print(rows_with_numeric_values)
       # Nümerik değerleri boşluk karakteri ile değiştirme
       data[target_column] =data[target_column].replace(to_replace='\d+(\.\d+)?', value=' ', regex=True)#ondalıklı sayıları da dahil etmek istiyorsak
    else:
        print(f"{target_column} sütunundaki tüm değerler metinsel.")
```

```python
#bu fonfksiyon verilen sütunda nümerik ifadeler var mı yok mu kontrol edilip varsa nümerik ifade yerine ' ' koymamızı sağlıyor.
numerical_value(target_column)
```

```python
numerical_value(target_column)
```

```python
def special_characters_query(target_column):
    rows_with_special_characters = data[data[target_column].str.contains('[^\w\s]', na=False)]
    # Eğer rows_with_special_characters boş değilse, o sütunda özel karakter içeren satırlar bulunuyor demektir.
    if not rows_with_special_characters.empty:
       print(f"{target_column} sütununda özel karakter içeren satırlar bulunuyor.")
       print(rows_with_special_characters)
    else:
       print(f"{target_column} sütunundaki tüm satırlar özel karakter içermiyor.")
```

```python
special_characters_query(target_column)
```

```python
# Noktalama işaretleri ve özel karakterleri temizleyen fonksiyon
def clean_text(text):
    # Noktalama işaretleri ve özel karakterleri temizleme
    cleaned_text = re.sub(r'[^\w\s]', '', str(text))
    return cleaned_text
```

```python
# Belirli sütunu temizleme
data[target_column] = data[target_column].apply(clean_text)
```

```python
# Önce nltk'nin stopwords'lerini indirin
nltk.download("stopwords")
# NLTK'nin punkt modelini indirme
nltk.download('punkt')
```

```python
# Gereksiz kelimeleri temizleyen fonksiyon
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(str(text))
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)
```

```python
# Belirli sütunu temizleme
data[target_column] = data[target_column].apply(remove_stopwords)
```

```python
data.head()
```

```python
#sadece body sütunundaki verileri küçük harfe çevirir
data['body'] = data['body'].apply(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
data.head()
```

```python
# "body" sütunundaki boş değerleri NaN yap
data[target_column].replace('', pd.NA, inplace=True)
```

```python
# NaN olan satırları sil
data.dropna(subset=['body'], inplace=True)
```

```python
data.reset_index(drop=True , inplace=True)
```

```python
data.drop( columns=['movie_name','Unnamed: 0','title'] , inplace=True)
```
